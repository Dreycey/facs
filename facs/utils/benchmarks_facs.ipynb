```python
import os
import matplotlib
import json
from pandas import DataFrame
from facs.utils import performance
```

```python
performance.fetch_couchdb_results()
```

```python
benchmarks = json.loads(performance.facs_vs_fastq_screen())
```

```python
print benchmarks.values()[0]
```

```python
frame = DataFrame(benchmarks.values())
```

```python
# Some minor preprocessing

frame.rename(columns={'delta': 'delta_fqscr'}, inplace=True)

frame['filter_facs'] = frame.filter_facs.apply(os.path.basename)
frame['sample_facs'] = frame.sample_facs.apply(os.path.basename)

# To compare facs filters with fast_screen reference indexes we normalize names
frame['filter_facs'] = [filt[0] for filt in frame.filter_facs.map(os.path.splitext)]

# Individual column with read counts
frame['reads'] = [os.path.splitext(reads.split('_')[-1])[0] for reads in frame.sample_facs]

# Normalize fastq_screen contamination rates
# XXX: to be removed soon since test_fastqscreen.py already normalizes this in new code
# frame['contam_fqscr'] = [ contam/100 for contam in frame.contam_fqscr ]
```

```python
accuracy = frame[ frame.sample_facs.str.contains('100vs400') ]
```

```python
accuracy
```

```python
accuracy = accuracy.loc[:,['contam_facs', 'contam_fqscr']]
accuracy.pivot_table(values=['contam_facs', 'contam_fqscr'], rows=['contam_fqscr'])
accuracy.plot(title='Contamination detection accuracy for simNGS-generated dm3 sample (~100% dm3)', kind='barh', logy=True, logx=True, figsize=(8,6))
```

```python
# Remove accuracy-only datasets so that they don't disturb in performance plots
# (operator ~ prepending the dataframe condition acts as "not" containing string)

frame = frame[ ~frame.sample_facs.str.contains('100vs400') ]
```

```python
# Slicing main dataframe into individual organism dataframes

dm3 = frame['sample_fqscr'].map(lambda x: 'dm3' in x)
eschColi = frame['sample_fqscr'].map(lambda x: 'eschColi_K12' in x)
phiX = frame['sample_fqscr'].map(lambda x: 'phiX' in x)

dm3 = frame[dm3]
eschColi = frame[eschColi]
phiX = frame[phiX]
```

```python
# Ready to plot

runtimes_eschColi = eschColi.loc[:,['reads', 'delta_fqscr', 'delta_facs']]
ecoli_table = runtimes_eschColi.pivot_table(values=['delta_fqscr', 'delta_facs'], rows=['reads'])
ecoli_table.plot(title='eschColi_K12 FACS vs Fastq_screen runtimes (log scale)', kind='bar', logy=True, logx=True)

runtimes_phiX = phiX.loc[:,['reads', 'delta_fqscr', 'delta_facs']]
phiX_table = runtimes_phiX.pivot_table(values=['delta_fqscr', 'delta_facs'], rows=['reads'])
phiX_table.plot(title ='phiX FACS vs Fastq_screen runtimes (log scale)', kind='bar', logy=True, logx=True)

runtimes_dm3 = dm3.loc[:,['reads', 'delta_fqscr', 'delta_facs']]
dm3_table = runtimes_dm3.pivot_table(values=['delta_fqscr', 'delta_facs'], rows=['reads'])
dm3_table.plot(title='eschColi_K12 FACS vs Fastq_screen runtimes (log scale)', kind='bar', logy=True, logx=True)
```
