```python
import os
import matplotlib
import json
from pandas import DataFrame
from facs.utils import performance
```

```python
performance.fetch_couchdb_results()
```

```python
benchmarks = json.loads(performance.facs_vs_fastq_screen())
```

```python
print benchmarks.values()[0]
```

```python
frame = DataFrame(benchmarks.values())
```

```python
# Some minor preprocessing

frame.rename(columns={'delta': 'delta_fqscr'}, inplace=True)

frame['filter_facs'] = frame.filter_facs.apply(os.path.basename)
frame['sample_facs'] = frame.sample_facs.apply(os.path.basename)

# To compare facs filters with fast_screen reference indexes we normalize names
frame['filter_facs'] = [filt[0] for filt in frame.filter_facs.map(os.path.splitext)]

# Individual column with read counts
frame['reads'] = [os.path.splitext(reads.split('_')[-1])[0] for reads in frame.sample_facs]
```
